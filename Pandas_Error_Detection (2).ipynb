```python
mdata=mdata.drop("YearBuilt",axis=1)
mdata
```

```python
for i in mdata.columns:
    if mdata[i].dtype in ["int64","float64"]:
        mdata[i].fillna(mdata[i].mean(),inplace=True)
    else:
        mdata[i].fillna(mdata[i].mode()[0],inplace=True)
```

```python
mdata.isna().sum()
```

```python
for x in mdata.columns:
    if mdata[x].dtype not in ["int64","float64"]:
        mdata=mdata.drop(x,axis=1)
        
        
```

```python
mdata=mdata.astype("int64")
```

```python
mdata
```

```python
mdata.dtypes
```

```python
plt.boxplot(mdata)
plt.show()
```

```python
for z in mdata.columns:
    sk=mdata[z].skew().round()
    if sk==0 or sk==-0:
        mean=mdata[z].mean()
        std1=mdata[z].std()
        lower=mean-(3*std1)
        upper=mean+(3*std1)
        mdata[z]=mdata[z].map(lambda a : lower if a < lower  else upper if a>upper else a)
    else:
        q1=mdata[z].quantile(0.25)
        q3=mdata[z].quantile(0.75)
        iqr=q3-q1
        lower=q1-(1.5*iqr)
        upper=q3+(1.5*iqr)
        mdata[z]=mdata[z].map(lambda t: lower if t < lower  else upper if t > upper else t)
```

```python
sns.boxplot(mdata)
plt.xticks(rotation=60)
plt.show()
```
